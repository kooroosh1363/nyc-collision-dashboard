```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#  read data set
df = pd.read_csv('/kaggle/input/vehicle-collisions/database.csv', low_memory=False)

# show data 
df.head()
```

## ðŸ•’ Temporal Analysis of Accidents

In this section, we explore when most accidents occur. We will:
- Extract date and time information
- Analyze accidents by hour, day of week, and month
- Create plots to visualize the distribution

```python
# Convert DATE and TIME columns to datetime
df['DATETIME'] = pd.to_datetime(df['DATE'] + ' ' + df['TIME'], errors='coerce')
```

```python
# Extract hour, day of week, and month from the datetime
df['HOUR'] = df['DATETIME'].dt.hour
df['WEEKDAY'] = df['DATETIME'].dt.day_name()
df['MONTH'] = df['DATETIME'].dt.month_name()
```

## ðŸ“Š Accident Frequency by Hour of Day

```python
# Plot distribution of accidents by hour
plt.figure(figsize=(12,6))
sns.countplot(x='HOUR', data=df, palette='viridis')
plt.title('Number of Collisions by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Collision Count')
plt.grid(True)
plt.show()
```

## ðŸ“† Accident Frequency by Day of Week


```python
# Set the order of days for plot
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

plt.figure(figsize=(12,6))
sns.countplot(x='WEEKDAY', data=df, order=day_order, palette='coolwarm')
plt.title('Number of Collisions by Day of Week')
plt.xlabel('Day')
plt.ylabel('Collision Count')
plt.grid(True)
plt.show()
```

## ðŸŒ Geospatial Analysis of Collisions

In this section, we analyze the geographic distribution of accidents.
We will:
- Check for missing coordinates
- Visualize points of accidents using scatterplots
- Identify spatial hotspots

```python
# Check how many rows have missing latitude or longitude
df[['LATITUDE', 'LONGITUDE']].isnull().sum()
```

```python
# Drop rows with missing LAT/LON
df_geo = df.dropna(subset=['LATITUDE', 'LONGITUDE'])
```

## ðŸ—ºï¸ Scatterplot of Collisions Across NYC

```python
# Plot all collisions on a scatterplot (rough map)
plt.figure(figsize=(10,10))
plt.scatter(df_geo['LONGITUDE'], df_geo['LATITUDE'], alpha=0.05, s=0.5, c='red')
plt.title('NYC Traffic Collisions - Geo Distribution')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

## ðŸ”¥ Heatmap of NYC Traffic Collisions

Now we'll use `folium` to create an interactive heatmap that shows areas with the highest collision densities in NYC.


```python
import folium
from folium.plugins import HeatMap

# Create a base map centered at NYC
m = folium.Map(location=[40.730610, -73.935242], zoom_start=11)

# Filter top 10000 rows for performance
heat_data = df_geo[['LATITUDE', 'LONGITUDE']].dropna().values[:10000]

# Add heatmap layer
HeatMap(heat_data, radius=8, blur=5).add_to(m)

# Show map
m
```

## ðŸš— Vehicle Type Analysis

We'll analyze which types of vehicles are most frequently involved in traffic collisions.  
The dataset includes up to 5 columns for vehicle types (`VEHICLE 1 TYPE`, ..., `VEHICLE 5 TYPE`).  
We'll focus on the first two for simplicity.


```python
# Top 10 most common vehicle types in VEHICLE 1 TYPE
df['VEHICLE 1 TYPE'].value_counts().head(10)
```

```python
# Plot top 10 vehicle types
top_veh1 = df['VEHICLE 1 TYPE'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_veh1.values, y=top_veh1.index, palette='mako')
plt.title('Top 10 Vehicle Types Involved in Collisions')
plt.xlabel('Number of Collisions')
plt.ylabel('Vehicle Type')
plt.grid(True)
plt.show()
```

## ðŸ§® Comparing VEHICLE 1 and VEHICLE 2

Now let's compare both `VEHICLE 1 TYPE` and `VEHICLE 2 TYPE` to see overall involvement of different vehicles.


```python
# Combine VEHICLE 1 and VEHICLE 2 columns
veh_all = pd.concat([
    df['VEHICLE 1 TYPE'],
    df['VEHICLE 2 TYPE']
], axis=0)

# Get top 15 most involved vehicle types
top_veh_all = veh_all.value_counts().head(15)

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=top_veh_all.values, y=top_veh_all.index, palette='flare')
plt.title('Most Involved Vehicle Types in All Collisions')
plt.xlabel('Total Appearances')
plt.ylabel('Vehicle Type')
plt.grid(True)
plt.show()
```

## ðŸ§  Contributing Factors Analysis

In this section, we analyze the most common causes of traffic collisions.  
We'll focus on the column `CONTRIBUTING FACTOR VEHICLE 1`, which indicates the main cause reported for the primary vehicle.


```python
# Top contributing factors for vehicle 1
df['VEHICLE 1 FACTOR'].value_counts().head(15)
```

```python
# Top 15 contributing factors
top_factors = df['VEHICLE 1 FACTOR'].value_counts().head(15)

plt.figure(figsize=(10,6))
sns.barplot(x=top_factors.values, y=top_factors.index, palette='crest')
plt.title('Top 15 Contributing Factors in Collisions (Vehicle 1)')
plt.xlabel('Number of Collisions')
plt.ylabel('Contributing Factor')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Combine both VEHICLE 1 and 2 factors
all_factors = pd.concat([
    df['VEHICLE 1 FACTOR'],
    df['VEHICLE 2 FACTOR']
], axis=0)

# Plot top combined causes
top_combined = all_factors.value_counts().head(15)

plt.figure(figsize=(10,6))
sns.barplot(x=top_combined.values, y=top_combined.index, palette='rocket')
plt.title('Top Contributing Factors from All Vehicles')
plt.xlabel('Collision Count')
plt.ylabel('Factor')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## ðŸ©¸ Injury & Fatality Analysis

Weâ€™ll analyze how many people were injured or killed in collisions, and who they were: pedestrians, cyclists, or motorists.

```python
# Total injuries and fatalities (correct column names)
injuries = df['PERSONS INJURED'].sum()
fatalities = df['PERSONS KILLED'].sum()

print(f'Total Injuries: {int(injuries):,}')
print(f'Total Fatalities: {int(fatalities):,}')
```

```python
# Breakdown of injuries and fatalities by group
group_columns = [
    'PEDESTRIANS INJURED',
    'PEDESTRIANS KILLED',
    'CYCLISTS INJURED',
    'CYCLISTS KILLED',
    'MOTORISTS INJURED',
    'MOTORISTS KILLED'
]

group_stats = df[group_columns].sum().sort_values(ascending=False)
print(group_stats)
```

```python
# Visualizing injuries and fatalities by group
plt.figure(figsize=(10,6))
sns.barplot(x=group_stats.values, y=group_stats.index, palette='Reds_r')
plt.title('Injuries and Fatalities by Group')
plt.xlabel('Total Count')
plt.ylabel('Group')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## ðŸ—ºï¸ Collision Analysis by NYC Boroughs

This section explores which boroughs in NYC have the highest number of collisions, injuries, and fatalities.
We'll use the `BOROUGH` column to group and aggregate data.

```python
# Total number of collisions per borough
df['BOROUGH'].value_counts()
```

```python
# Plot number of collisions by borough
plt.figure(figsize=(10,6))
sns.countplot(x='BOROUGH', data=df, order=df['BOROUGH'].value_counts().index, palette='Set2')
plt.title('Number of Collisions per Borough')
plt.xlabel('Borough')
plt.ylabel('Collision Count')
plt.grid(True)
plt.show()
```

```python
# Grouping injuries and fatalities by borough
borough_stats = df.groupby('BOROUGH')[['PERSONS INJURED', 'PERSONS KILLED']].sum().sort_values(by='PERSONS INJURED', ascending=False)
borough_stats
```

```python
# Barplot for injuries and fatalities by borough
borough_stats.plot(kind='bar', figsize=(10,6), colormap='coolwarm')
plt.title('Injuries and Fatalities by Borough')
plt.xlabel('Borough')
plt.ylabel('Count')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
